$Step-1$ import the packages

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

$Step-2$ Read the data

$Step-3$ Divide data input and output

```python
X=df.drop('Customer Status',axis=1)
y=df['Customer Status']
```

```python
X
```

```python
y
```

$Step-4$ Train test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

$Step-5$ Read the base model

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeClassifier
grid_tree=DecisionTreeClassifier()#Base model
grid_tree
```

$Step-6$ Create a Parameter file

```python
grid_tree.get_params()
```

```python
param_grid = {
             'criterion':['gini','entropy'],
             'max_depth':[3,4,5,6,7,8],
             'min_samples_split':[2,3,4],
             'min_samples_leaf':[1,2,3,4],
             'random_state':[42,1234]
             }
#2*6*3*4*2=288
```

$Step-7$ Aplly Grid Search CV

```python
import time
start=time.time()
grid_search=GridSearchCV(grid_tree,param_grid,scoring='accuracy',cv=5,verbose=True)

end=time.time()
print("Total time taken is:",(end-start))
```

```python
grid_search
```

$Step-8$  Fit the data

```python
import time
start=time.time()
grid_search.fit(X_train,y_train)
end=time.time()
print("Total time taken is:",(end-start))
```

$Step-9$ Get the Best Parameters

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

Cross validation Score

```python
accuarcy_list=cross_val_score(grid_search.best_estimator_,X_train,y_train,scoring='accuracy')
accuarcy_list
```

```python
accuarcy_list.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy',
                             max_depth= 7,
                             min_samples_leaf= 4,
                             min_samples_split= 2,
                             random_state= 42)
dtree.fit(X_train,y_train)
```

```python
y_pred_dt=dtree.predict(X_test)
```

```python
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score
acc_dt=round(accuracy_score(y_test,y_pred_dt)*100,2)
precision_dt=round(precision_score(y_test,y_pred_dt),2)
recall_dt=round(recall_score(y_test,y_pred_dt),2)
f1_dt=round(f1_score(y_test,y_pred_dt),2)
print("Accuracy",acc_dt)
print("precision Score",precision_dt)
print("Recall score",recall_dt)
print("F1 Score",f1_dt)
```
